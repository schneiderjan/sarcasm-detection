```python
import pandas as pd
import numpy as np

warriner_ratings = pd.read_csv("data/Ratings_Warriner_et_al.csv")
```

```python
columns = warriner_ratings.columns
print(columns)
```

```python
means_sums = warriner_ratings[['Word', 'V.Mean.Sum',
       'A.Mean.Sum','D.Mean.Sum',]]
means_sums
```

```python
import nltk, re, pprint
from nltk.tokenize.casual import TweetTokenizer
tknzr = TweetTokenizer()
t1= "'Creativity Defined' at Marketing stage talks the importance of innovation and passion, with amazing insight."
t2= "zoom zoning abort abortion absent"
```

```python
def extract_subtracted_ratings(ratings):
    subValence = ratings["V.Mean.Sum"].max() - ratings["V.Mean.Sum"].min()
    subArousal = ratings["A.Mean.Sum"].max() - ratings["A.Mean.Sum"].min()
    subDominance = ratings["D.Mean.Sum"].max() - ratings["D.Mean.Sum"].min()
    subtractedRatings = [subValence, subArousal, subDominance]
    return subtractedRatings  
```

```python
def assign_rating_to_tweet(tweet):
    columns=['Word','V.Mean.Sum','A.Mean.Sum','D.Mean.Sum']
    unigrams = tknzr.tokenize(tweet)
    index = range(len(unigrams))
    ratings = pd.DataFrame(columns=columns, index=index)
    
    for i in range(len(unigrams)):
        df = means_sums.loc[means_sums['Wordvalues'] == unigrams[i].lower()]
        ratings.loc[i] = df.
        
    return extract_subtracted_ratings(ratings)
```
