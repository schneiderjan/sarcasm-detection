```python
df_ama = pd.read_csv("data/sentiment_labelled_sentences/amazon_cells_labelled.txt", sep="\t", header=None)
df_ama.columns=["text","target"]
df_ama.head(4)
```

```python
df_imdb = pd.read_csv("data/sentiment_labelled_sentences/imdb_labelled.txt", sep="\t", header=None)
df_imdb.columns=["text","target"]
```

```python
print(df_imdb["target"].isnull().sum())
df_imdb.head(4)
```

```python
df_yelp = pd.read_csv("data/sentiment_labelled_sentences/yelp_labelled.txt", sep="\t", header=None)
df_yelp.columns=["text","label"]
df_yelp.head(4)
```

```python
df_yelp["label"].isnull().sum()
```

```python
df_yelp = df_yelp.dropna()
```

```python
df_ama.append(df_yelp)
df_ama.append(df_imdb)
```

```python
df_ama["target"].value_counts()
```

```python
check1 = [1]
check0 = [0]

df_ama["target"] = df_ama["target"].astype(int)

df_pos = df_ama[df_ama["target"].isin(check1)]
df_neg = df_ama[df_ama["target"].isin(check0)]
```

```python
df_pos.to_csv("data/positive_dataset.csv", header=["text","target"],index=False)
df_neg.to_csv("data/negative_dataset.csv",header=["text","target"],index=False)
```
