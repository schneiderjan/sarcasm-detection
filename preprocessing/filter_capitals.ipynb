# 2. Capitals

```python
#Read data from csv
df = pd.read_csv("data/tweets_small.csv", delimiter='|', encoding="utf-8", header=None, names=['TweetID', 'Tweet','Status'])

#Remove if empty
df = df.dropna()

#Pre-process tweet & split individual words
new_tweet = df['Tweet']
new_tweet = new_tweet.str.replace(',', '').str.replace('.', '')
new_tweet = new_tweet.str.replace('!', '').str.replace('?', '')
new_tweet = new_tweet.str.replace('/', '').str.replace('|', '')
new_tweet = new_tweet.str.replace('^', '')
new_tweet = new_tweet.str.strip('"')

#Check for CAPITAL words & put them in array
new_tweet = new_tweet.str.split()
capitals = new_tweet.apply(lambda words: [word for word in words if word.isupper()])
#df['CW'] = capitals

count = 0
for l in capitals :
    if len(l) > 0 :
        ctf = 1
    else :
        ctf = 0
    df.loc[count,'CTF'] = ctf
    count+=1

#Display datafram
df
```
